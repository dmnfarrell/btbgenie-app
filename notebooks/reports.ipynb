```python
def report(cent, parcels, moves=None, level='snp3', clades=[], cmap='tab20'):
    """Generate html and pdf report for cluster"""
    
    from weasyprint import HTML
    sub = cent[cent[level].isin(clades)].copy()    
    herds = list(sub.HERD_NO)
    p = parcels[parcels.SPH_HERD_N.isin(herds)]    
    cols = ['Animal_ID','HERD_NO']#,'move_from','move_date','data_type']
    
    fig,ax=plt.subplots(1,1)    
    parcels.plot(column='SPH_HERD_N',alpha=0.6,lw=1,cmap=cmap,ax=ax)
    if moves is not None:
        mov = gui.get_moves_bytag(sub, moves, lpis_cent)
        gui.plot_moves(mov, lpis_cent, ax=ax)
    gui.plot_single_cluster(sub,col=level,ax=ax)
    fig.suptitle('%s=%s n=%s' %(level,clades,len(sub)))
    plt.tight_layout()
    tempname = tempfile.mktemp()
    fig.savefig(tempname, format='png', dpi=150)

    css_file = 'style.css'
    s = '<html>'
    s += '<head><link rel="stylesheet" href="%s"><head>' %css_file
    s += '<h1>cluster report</h1>'
    s += '<p>%s</p>' %datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    s += '<div><img src="%s" class="small-scaled"></div>' % tempname
    table = '<div class="container"><table> %s</table><div>' %sub[cols].reset_index().to_html()
    s+=table

    with open('report.html', 'w') as f:
        f.write(s)
   
    #html = HTML(string=s)
    html = HTML(string=open('report.html', 'rb').read(), base_url='./')
    
    html.write_pdf('test.pdf', stylesheets=[css_file], 
                   optimize_images=True, jpeg_quality=70, dpi=150)
    
report(cent, parcels, moves, clades=['91'])
```
