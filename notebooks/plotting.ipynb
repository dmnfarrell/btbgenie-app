# Plot methods for tracebtb

Uses genotype distance from SNP analysis to correlate and score risk factors using movement data within herds.

## Keys:

* **all_moves_from_not_sl** shows all cases where one of the Monaghan sample animlas (N = 112) moved from a herd during its lift, excluding its final move to slaughter. Moves from herd with no record of a breakdown are shown with NULLs or 0s in all columns which measure the existence of breakdowns in those herds
* **summary_move_from** shows a summary of the "all_moves_from_not_sl"  dataset, summarised for each of the sample animals. Cases where the animal has not moved from any herd until its slaughter data are indicted with NULLS in most of the columns.
* **period moves** shows all time period where one of the Monaghan sample animals (N = 112) was in a herd which experienced a breakdown at some time, as well as all periods where the animal was in a herd with another animal (the "matched animal") and whether the matched animal had itself been in any herd which experienced a breakdown at any time.
* **summary periods** shows a summary of the "period moves"  dataset, summarised for each of the sample animals. Cases where the animal has not moved from any herd until its slaughter data are indicted with NULLS in most the columns.

## Methods:

* find movement props for outlier clade samples - may not be 'endemic' strains 
    - are they more likely to have moved in recently or contacted with moves from same herd?
* isolates in same herd - look at clade or distance, distance match movements?
* look at all isolates with recent moves and/or contacts in herd and plot on phylogram?

```python
def kde_plot(gdf, p, color='#507CBD', percentiles=[10, 20, 40, 60, 75, 95], alpha=0.5):
    """KDE plot with specified percentiles."""

    # Perform the KDE computation
    x, y, z = bokeh_plot.kde(gdf, 200)
    
    # Flatten z to 1D and sort for cumulative density calculation
    z_flat = z.ravel()
    z_sorted = np.sort(z_flat)
    cumulative_density = np.cumsum(z_sorted)

    # Normalize cumulative density to range [0, 1]
    cumulative_density /= cumulative_density[-1]

    # Convert percentile levels to a range [0, 1] for cumulative density matching
    percentile_levels = np.array(percentiles) / 100.0

    # Get the corresponding z values for each specified percentile
    lvl = np.interp(percentile_levels, cumulative_density, z_sorted)
   

    # Plot KDE contours at the specified percentiles
    p.grid.level = "overlay"
    palette = bokeh_plot.create_custom_palette(color, num_colors=len(lvl))
    
    # Draw contours corresponding to the percentile levels
    p.contour(x, y, z, lvl, fill_color=palette, line_color=palette, fill_alpha=alpha)
    
    return
```

```python
def spatial_cluster(gdf,ax=None):
    """Spatially cluster points"""
    
    import sklearn.cluster as skc
    coordinates = gdf['geometry'].apply(lambda p: np.hstack(p.xy)).values
    coordinates = np.vstack(coordinates)
    clusterer = skc.DBSCAN(eps=3000).fit(coordinates)
    nclusters = clusterer.p
    gdf = gdf.assign(cl=clusterer.labels_)
    #print (gdf)
    if ax is not None:
        gdf.plot('cl', k=nclusters, cmap='tab20', ax=ax)#, legend=True)
        ax.axis('off')
    return gdf  

#sub=kilk
print (len(sub))
fig,ax=plt.subplots(1,2,figsize=(12,10))
cl = spatial_cluster(sub,ax[0])

sub['color'],c = tools.get_color_mapping(sub, 'snp20')
gui.plot_single_cluster(sub, ms=40, margin=1000,
                        legend=False, title='', ax=ax[1])
```

```python
def hexbin(gdf, n_bins=10, p=None):
    """Hex bin plot of point counts"""
    
    from bokeh.transform import linear_cmap
    from bokeh.palettes import Blues256, Reds256
    from bokeh.util.hex import hexbin
    
    if p is None:
        p = bokeh_plot.init_figure(provider='CartoDB Positron')   
        
    gdf = gdf.to_crs('EPSG:3857') 
    gdf['x'] = gdf.geometry.x
    gdf['y'] = gdf.geometry.y   
    x = gdf['x'].values
    y = gdf['y'].values
    
    width = x.max() - x.min()
    hex_size = width/n_bins
    print (width,hex_size)
    bins = hexbin(x, y, hex_size)   
    source = ColumnDataSource(data=dict(
        q=bins.q, r=bins.r, counts=bins.counts
    ))
    #color mapper for coloring hexagons based on point counts
    color_mapper = linear_cmap(field_name='counts', palette=Reds256[::-1], low=min(bins.counts), high=max(bins.counts))
    tiles = p.hex_tile(q="q", r="r", size=hex_size, line_color='black', source=source, 
               fill_alpha=0.7, fill_color=color_mapper)
    hover = HoverTool(
        tooltips=[("Count", "@counts")],  # Display the 'counts' field from ColumnDataSource
        mode="mouse",  # Display hover info wherever the mouse is over a tile
        renderers=[tiles]  # Apply hover tool to hex tiles
    )
    p.add_tools(hover)    
    return p

animal = meta[meta['sample']=='TB22-001491'].iloc[0]
point = animal.geometry
sub = meta[meta.geometry.distance(point) <= 90000].copy()
p=hexbin(sub)
show(column(p,width=600,height=600))
```

```python
reload(bokeh_plot)

def scatter_pie(gdf, groupby='HERD_NO', col='snp12', colormap=None,
                radius=2000, legend=True, legend_fontsize=12, provider='CartoDB Positron', p=None):
    """Draw wedges colored by proportion of points in a group"""

    from sklearn.cluster import KMeans
    from bokeh.transform import cumsum
    from bokeh.models import Wedge

    gdf = gdf.to_crs('EPSG:3857')
    if p == None:
        p = bokeh_plot.init_figure('', provider)
    #if groupby == 'cl':
    #    gdf = tools.spatial_cluster(gdf,2000)
    groups = gdf.groupby(groupby)

    rend = []
    for herd, group in groups:
        if len(group) <= 1:
            continue
        geo_source = GeoJSONDataSource(geojson=group.to_json())
        pt = group.union_all().centroid
        x, y = pt.x,pt.y

        #summarize group
        summary = group.groupby(col).size()
        summary = summary.reset_index(name='value').sort_values(by=col)
        summary['angle'] = summary['value'] / summary['value'].sum() * 2 * np.pi
        summary['color'] = summary[col].map(colormap)
        summary['radius'] = len(group)*200+200
        summary['size'] = len(group)
        row = group.iloc[0]
        summary['herd'] = row.HERD_NO

        #print (summary)
        source = ColumnDataSource(summary)

        r = p.wedge(x=x, y=y, radius='radius', source=source,
                start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
                color='color', line_color="black", fill_alpha=0.5)
        rend.append(r)

    h = HoverTool(renderers=rend, tooltips=([(col, f"@{col}"),
                                            ('herd',"@herd"),
                                            ('size',"@size")]))
    p.add_tools(h)
    if legend == True and col != None:
        bokeh_plot.add_legend(p, gdf, col, legend_fontsize)
    p.axis.visible = False
    p.toolbar.logo = None
    return p
    
col='snp50'
sub['color'],cm = tools.get_color_mapping(sub, col)

reload(bokeh_plot)
reload(tools)
#print (sub)
#p1 = bokeh_plot.plot_selection(sub, col=col, legend=True)
p2 = scatter_pie(sub, 'HERD_NO', col, cm)
# Show plot
show(row(p2,width=800,height=600))
```

```python
#get some moves randomly over a timescale
allmov

from matplotlib.animation import FuncAnimation
import numpy as np

# Create a function that updates the plot for each frame of the animation
def animate(i):
    # Clear the current plot
    plt.clf()
    
    # Generate some data to plot (you can replace this with your own data)
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(x + i * 0.1)  # Update the data for animation

    # Plot the data
    plt.plot(x, y, label='Sine Wave')
    plt.title('Animating a Sine Wave')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()

# Create a figure and an axes object
fig, ax = plt.subplots()

# Create an animation object
ani = FuncAnimation(fig, animate, frames=50, repeat=False, blit=False)

# Display the animation (you may need to install ffmpeg to save the animation as a video)
plt.show()
```
